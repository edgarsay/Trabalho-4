# K-means

```python
class KMeans:
    def __init__(self, K=2, max_iter=30, thresh_hold=-1) -> None:
        self.K = K
        self.centroids = [None for _ in range(K)]
        self.centroids_label = [None for _ in range(K)]
        self._pre_centroids = [None for _ in range(K)]
        self.clusters = [ [] for _ in range(K)]
        self.max_iter = max_iter
        self.thresh_hold = thresh_hold
    
    def _update_groups(self, X_values:list, y:list):
        self.clusters = [ [] for _ in range(self.K)]
        for i in range(len(X_values)):
            x_value = X_values[i]
            y_value = y[i]
            # find the closest cluster
            closest_index = find_closest_neighbor(x_value, self.centroids)
            # put the value in the closest cluster's group
            self.clusters[closest_index].append([*x_value, y_value])

    def _update_clusters_label(self):
        self.centroids_label = [None for _ in range(self.K)]
        for i in range(self.K):
            self.centroids_label[i] = compute_moda(
                list(map(lambda e: e[-1], self.clusters[i]))
            )

    def fit(self, X:pd.DataFrame, y:list):
        X_values = X.values
        # set the init clusters at a random elements position
        for i in range(self.K):
            self.centroids[i] = random.choice(X.values)
            self._pre_centroids[i] = self.centroids[i]
        for _ in range(self.max_iter):
            # group the element using the closest cluster
            self._update_groups(X_values, y)
            self._update_clusters_label()
            # compute the centroid of each group
            new_position = -1
            for index, group in enumerate(self.clusters):
                # store previous cluster position
                self._pre_centroids[index] = self.centroids[index]
                # update clusters's center values
                group_values = list(map(lambda e: np.array(e[:-1]), group))
                if len(group_values) > 0:
                    new_position = compute_mean(group_values)
                    self.centroids[index] = new_position
            # update group to the new positions
            self._update_groups(X_values, y)
            self._update_clusters_label()
            # if there's a valid thresh_hold value
            if self.thresh_hold > 0:
                # Check if distance from the pre position is bigger
                # than the thresh_hold, for all clusters
                all_below_thresh_hold = True
                for index in range(self.K):
                    crr_clu = self.centroids[i]
                    pre_clu = self._pre_centroids[i]
                    distance = euclidean_distance(crr_clu, pre_clu)
                    if distance > self.thresh_hold:
                        all_below_thresh_hold = False
                        break
                # if so, stop the iteractions loop
                if all_below_thresh_hold:
                    break
        # Remove unused centroid, _pre_centrois and cluster
        i = 0
        while i < len(self.clusters):
            if len(self.clusters[i]) == 0:
                self.centroids.pop(i)
                self.centroids_label.pop(i)
                self._pre_centroids.pop(i)
                self.clusters.pop(i)
                continue
            i += 1

    def predict(self, X:pd.DataFrame):
        predicted_labels = []
        for value in X.values:
            closest_i = find_closest_neighbor(value, self.centroids)
            predicted_labels.append(self.centroids_label[closest_i])
        return predicted_labels

    @staticmethod
    def compute_Acc_Std(dataset, label_column, u_labels, k, n_int=20):
        cms = []
        for _ in range(n_int):
            model = KMeans(K=k, max_iter=30)
            pred, actu = interaction(dataset, model, label=label_column)
            cm = confusion_matrix(u_labels, actu, pred)
            cms.append(np.array(cm))
        # compute hit rates
        hit_rates = compute_hit_rates(cms)
        # compute accuracy and standar_deviation
        acc = compute_mean(hit_rates)
        std = compute_standard_deviation(hit_rates)
        return acc, std
    
    @staticmethod
    def compute_k_score(k, acc, std, max_k):
        return acc - std - (k / (max_k * 10))

    @staticmethod 
    def lg_Ks_Accs_Stds(dataset, label_column, max_k):
        Ks, Accs, Stds = [], [], []
        n_interactions = 20
        u_labels = list(dataset[label_column].unique())
        start = 1
        end = max_k
        acc, std = KMeans.compute_Acc_Std(dataset, label_column, u_labels, start)
        Ks.append(start)
        Accs.append(acc)
        Stds.append(std)
        score_start = KMeans.compute_k_score(start, acc, std, max_k)
        acc, std = KMeans.compute_Acc_Std(dataset, label_column, u_labels, end)
        Ks.append(end)
        Accs.append(acc)
        Stds.append(std)
        score_end = KMeans.compute_k_score(end, acc, std, max_k)
        mid = int(((end - start) / 2) + start)
        for _ in tqdm(range(int(math.log2(max_k))+1)):
            if start == mid:
                acc, std = KMeans.compute_Acc_Std(dataset, label_column, u_labels, start)
                Ks.append(start)
                Accs.append(acc)
                Stds.append(std)
                score_start = KMeans.compute_k_score(start, acc, std, max_k)
            elif end == mid:
                acc, std = KMeans.compute_Acc_Std(dataset, label_column, u_labels, end)
                Ks.append(end)
                Accs.append(acc)
                Stds.append(std)
                score_end = KMeans.compute_k_score(end, acc, std, max_k)

            mid = int(((end - start) / 2) + start)
            if score_start > score_end:
                end = mid
            else:
                start = mid
        return Ks, Accs, Stds

    @staticmethod
    def get_the_best_k(Ks:list, Accs:list, Stds:list):
        max_k = max(Ks)
        best_k = Ks[0]
        best_k_score = KMeans.compute_k_score(best_k, Accs[0], Stds[0], max_k)
        for i in range(1, len(Ks)):
            crr_k_score = KMeans.compute_k_score(Ks[i], Accs[i], Stds[i], max_k)
            if crr_k_score >= best_k_score:
                best_k_score = crr_k_score
                best_k = Ks[i]
        return best_k
```

# Iris Dataset

```python
# Download Iris data
download_file('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
```

```python
# Load Iris dataset as a pandas DataFrame
iris_column_names = ['Sepal_Length', 'Sepal_Width','Petal_Length', 'Petal_Width', 'Class']
iris = pd.read_csv('./datasets/iris.data', names=iris_column_names, sep=',')
iris.head(2)
```

```python
# Does interactions over the Iris dataset, colecting accuracy and standard_deviation
dataset = iris
label_column = 'Class'
train_size = int(.8 * len(dataset))
max_k = train_size
file_name = './K_means_cache/lg_iris_Ks_Accs_Stds.csv'
try:
  # if alread done
  Ks_Accs_Stds = pd.read_csv(file_name)
  Ks = Ks_Accs_Stds['Ks']
  Accs = Ks_Accs_Stds['Accs']
  Stds = Ks_Accs_Stds['Stds']
except:
  # else not, compute now
  Ks, Accs, Stds = KMeans.lg_Ks_Accs_Stds(
    dataset,
    label_column,
    max_k
  )
  # then save the result
  pd.DataFrame({
    'Ks': Ks,
    'Accs': Accs,
    'Stds': Stds,
  }).to_csv(file_name)
```

```python
# Get the best k, with the high accuracy and small standard deviation
best_k = KMeans.get_the_best_k(Ks, Accs, Stds)
print('Best K: ', best_k)
```

```python
# Compute a confusion matrix or "matriz de confusão"
plot_acc_std_closest_cm_to_the_mean(iris, 'Class', KMeans, [best_k], 'K-Means (k={})'.format(best_k))
```

```python
# Generate dataset, Artificial II
dataset_size = 150
noise = 0.1
centroid_size = int(dataset_size / 3)
classe1_size = centroid_size
classe2_size = centroid_size
classe3_size = centroid_size
c1_centroids = [
    [0.25, 0.75],
]
c2_centroids = [
    [0.75, 0.75],
]
c3_centroids = [
    [0.50, 0.25],
]
x1, x2, label = [], [], []
for centroid in c1_centroids:
    for _ in range(classe1_size):
        noise1 = random.random() * noise - noise / 2
        noise2 = random.random() * noise - noise / 2
        x1.append(centroid[0] + noise1)
        x2.append(centroid[1] + noise2)
        label.append('classe_1')
for centroid in c2_centroids:
    for _ in range(classe2_size):
        noise1 = random.random() * noise - noise / 2
        noise2 = random.random() * noise - noise / 2
        x1.append(centroid[0] + noise1)
        x2.append(centroid[1] + noise2)
        label.append('classe_2')
for centroid in c3_centroids:
    for _ in range(classe3_size):
        noise1 = random.random() * noise - noise / 2
        noise2 = random.random() * noise - noise / 2
        x1.append(centroid[0] + noise1)
        x2.append(centroid[1] + noise2)
        label.append('classe_3')
```
